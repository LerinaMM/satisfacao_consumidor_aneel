## 3.2 Alterando os Tipos das Colunas

```python
# Selecionando as colunas do tipo object
cols_obj = df3.select_dtypes(include = ['object']).columns

# Observando os tipos de dados das colunas do tipo object
for col in cols_obj:
    print(f'Coluna: {col:<50} Tipos de dados: {df3[col].apply(type).unique()}')
    # :<50 = serve para aumentar o espaçamento horizontal entre o texto
```

```python
# Criando um loop para tentar converter os dados em float
for col in cols_obj:
    try:
        df3[col] = df3[col].astype(float)

# Em caso de erro, o loop vai tentar substituir a "," por "." e vai tentar novamente converter para float
    except ValueError:
        try:
            df3[col] = df3[col].str.replace(',', '.').astype(float)
            df3[col] = df3[col].astype(float)

# Em caso de erro novamente, o loop vai parar e informar que não foi possível converter a coluna
        except ValueError:
            print(f"Não foi possível converter a coluna {col} para float. Pulando esta coluna.")
    continue
```

## 4.3 Alterando o Tipo de Capital de Acordo com o Ano de Privatização

```python
df_merged['Tipo'] = df_merged.apply(lambda x: 'Privada' if (x['NumAno'] > x['AnoPrivatizacao']) & (x['AnoPrivatizacao'] != 0) else x['Tipo'], axis = 1)
df_merged[['NumAno','SigAgente','Tipo']]
```

## 4.4 Preenchendo NAs

```python
for col in df_merged.columns[6:10]:    
    
    # Média entre os valores posteriores e anteriores ao valor NaN
    rolling_mean = df_merged[col].rolling(window = 2, min_periods = 1).mean()

    # Preencha os valores NaN com a média
    df_merged[col].fillna(rolling_mean, inplace=True)
    
df_merged.isna().sum()
```

## 4.5 Adicionando Atributo que Calcula a Média Entre os Indicadores

```python
df_merged['MediaIndicadores'] = df_merged.apply(lambda x: 
                                                round((x['MdaIndicadorQualidade'] + x['MdaIndicadorSatisfacao'] +
                                                x['MdaIndicadorFidelidade'] + x['MdaIndicadorValor'] +
                                                x['MdaIndicadorConfianca'])/5,2), axis = 1)
df_merged['MediaIndicadores']
```

## 5.1 Descrição do Problema Analisado

**Fontes dos dados:**
- Dados dos índices de satisfação: https://dadosabertos.aneel.gov.br/dataset/indice-aneel-de-satisfacao-do-consumidor-iasc/resource/7abd3a47-1e1a-4a33-933b-48d658a6f912
- Unidades federativas e unidades consumidoras: https://app.powerbi.com/view?r=eyJrIjoiNDI4ODJiODctYTUyYS00OTgxLWE4MzktMDczYTlmMDU0ODYxIiwidCI6IjQwZDZmOWI4LWVjYTctNDZhMi05MmQ0LWVhNGU5YzAxNzBlMSIsImMiOjR9&pageName=ReportSection
- Privatizações: (1) https://www.andes.org.br/conteudos/noticia/eletrobras-governo-privatiza-a-maior-empresa-de-energia-eletrica-da-america-latina1, (2) https://www.correiobraziliense.com.br/economia/2021/01/4898127-agenda-de-privatizacoes-desestatizacao-no-setor-eletrico-continuara-este-ano.html

**Dicionário dos indicadores:**
- **Indicador de Qualidade**: leva em consideração a qualidade das informações passadas aos consumidores por parte das distribuidoras, o atendimento igualitário aos consumidores, segurança no valor cobrado, facilidade de acesso aos canais da distribuidora, Pontualidade, velocidade e cordialidade na prestação do serviço, fornecimento de energia sem interrupção e sem variação de tensão, avisos antecipados sobre cortes de energia e sobre falta de energia devido a manutenções.
- **Indicador de Satisfação**: leva em consideração a satisfação geral dos consumidores.
- **Indicador de Fidelidade**: leva em consideração se o consumidor mudaria de distribuidora por mais qualidade, mas com preço 20% maior e se indicaria a distribuidora para parentes e amigos.
- **Indicador de Valor**: levando em considerando a qualidade, avalia o preço e levando em consideração o preço, avalia a qualidade. Valor percebido ou custo-benefício.
- **Indicador de Confiança**: leva em consideração a confiança geral dos consumidores nas distribuidoras e a preocupação com os consumidores.

**Premissas:**
- Como as concessionárias atendem cerca de 99% da população brasileira, foram retiradas das análises as permissionárias (207.285.519 pessoas das concessionárias vs 1.203.328 pessoas das permissionárias - dados de 02/09/2023).
- Para concessionárias que atendem em mais de um estado foi considerado na análise apenas o estado com mais unidades consumidoras atendidas.
- Para concessionárias estatais ou privatizadas antes de 2006 o ano de privatização foi preenchido com 0.
- Para algumas concessionárias, a quantidade de unidades consumidoras atendidas encontrada foi aproximada e não exata (dados das unidades consumidoras coletados no dia 02/09/2023).
- O índice de fidelidade referente ao ano de 2020 não foi coletado pela ANEEL. Deste modo, o valor foi preenchido com a média entre os índices dos anos imediatamente anteriores e posteriores (2019 e 2021), para assim diminuir o máximo possível o efeito sobre o cálculo do valor médio entre os índices.
- Nota explicativa: a ANEEL esclarece que a pesquisa realizada em 2011 não teve os resultados divulgados em decorrência da não validação da etapa de campo.

## 5.2 Correlações

```python
df5 = df_final.copy()

df5['tipo'] = df5['tipo'].apply(lambda x: 0 if x == 'Estatal' else 1)

# Selecionando apenas os atributos numéricos para aplicação do método de Pearson
num_attributes = df5.select_dtypes(include = ['int64', 'float64'])
num_attributes = num_attributes.drop(['id', 'ano_privatizacao'], axis = 1)

correlation = num_attributes.corr(method = 'pearson')
ax = sns.heatmap(correlation, annot = True, annot_kws = {"size": 20});
ax.set_yticklabels(ax.get_yticklabels(), size = 18);
ax.set_xticklabels(ax.get_xticklabels(), size = 12);
```
